```python
df['longitude'].fillna(df['longitude'].mean(),inplace=True)
df['latitude'].fillna(df['latitude'].mean(),inplace=True)
df['elevation'].fillna(df['elevation'].mean(),inplace=True)
df['lot_acres'].fillna(df['lot_acres'].mean(),inplace=True)
```

```python
df['longitude'].isnull().unique()
```

```python
for column in df.columns.values.tolist():
    df_0=(df[column] == 0).sum()+(df[column].isna()).sum()
    print(column,df_0)
```

```python
df['rooms'].replace((0,(df['rooms']>0).mean()),inplace=True)
df['beds'].replace((0,(df['beds']>0).mean()),inplace=True)
df['year_built'].replace((0,(df['year_built']>0).mean()),inplace=True)
```
