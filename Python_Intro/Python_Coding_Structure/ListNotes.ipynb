## add & del elements into/from list ##
#### 5 adding methods ####
当涉及列表增加和删除元素时，列表会自动进行内存管理，大大减少程序员的负担。但这回涉及元素的大量移动，侵占运算量，效率低。除非必要我们一般只在列表的微服添加元素

```python
## .append() add new elements to the end of list; the fastest way and be recommonded
a=[20,49]
print("the id of a:",id(a))
a.append(200)
print(a) 
print("the id of a:",id(a))
```

```python
## +[] instead of adding elements to orginal list, 
## "+" copy&pasted 2 lists together and form a new object 
a=[20,49]
print(a)
print("the id of a:",id(a))
a=a+[200]
print(a)
print("the id of a:",id(a))
```

they have different id -> **formed a new object** 

```python
## extend[]  samiliar as append. doesn't form a new object  
a=[20,40]
print("the id of a:",id(a))
a.extend([200])
print(a)
print("the id of a:",id(a))
```

```python
## insert(position,number) insert a specific element into any position of this list
a=[10,20,30,40]
print("the id of a:",id(a))
a.insert(2,100)
print(a)
print("the id of a:",id(a))
a.insert(2,[50,60]) ## insert() can only insert 1 element 
print(a)
```

```python
def if_function(condition, true_result, false_result):
    """Return true_result if condition is a true value, and
    false_result otherwise.

    >>> if_function(True, 2, 3)
    2
    >>> if_function(False, 2, 3)
    3
    >>> if_function(3==2, 3+2, 3-2)
    1
    >>> if_function(3>2, 3+2, 3-2)
    5
    """
    if condition:
        return true_result
    else:
        return false_result


def with_if_statement():
    """
    >>> result = with_if_statement()
    47
    >>> print(result)
    None
    """
    if cond():
        return true_func()
    else:
        return false_func()

def with_if_function():
    """
    >>> result = with_if_function()
    42
    47
    >>> print(result)
    None
    """
    return if_function(cond(), true_func(), false_func())

def cond():
    return False

def true_func():
    print(42)

def false_func():
    print(47)
```
